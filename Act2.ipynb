### Transfer Function Definition
G(s) = 10(s+2)/(s(s+5)(s+10))

```python
import control as ctrl
import matplotlib.pyplot as plt

# Define the numerator and denominator coefficients
num = [10, 20]
den = [1, 15, 50, 0]

G = ctrl.tf(num, den)
print(G)
```

```python
t, y = ctrl.step_response(G)
plt.plot(t, y)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```
